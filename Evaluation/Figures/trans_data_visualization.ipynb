```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

path = "/home/yuchen/Crickets_Walking_Motion_Prediction/Evaluation/Figures/trans_eval_boxplot.csv"
data = pd.read_csv(path, header=0, usecols=[0,1,2])
data

# Usual boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='model', y='vel',  hue="step", data=data, width=0.6, showmeans=True, palette="blend:#FFDFDF,#FF8884")
plt.xticks(fontsize=14)  
plt.yticks(fontsize=14)
plt.xlabel('Model',fontsize=14)
plt.ylabel('Vel [mm/s]',fontsize=14)
plt.title('Multi-step Velocity Prediction (RMSE)',fontsize=14)
plt.legend(fontsize=14)
plt.show()

path = "/home/yuchen/Crickets_Walking_Motion_Prediction/Evaluation/Figures/trans_eval_boxplot.csv"
data = pd.read_csv(path, header=0, usecols=[0,1,3])
data

# Usual boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='model', y='direction',  hue="step", data=data, width=0.6, showmeans=True, palette="blend:#FFDFDF,#FF8884")
plt.xticks(fontsize=14)  
plt.yticks(fontsize=14)
plt.xlabel('Model',fontsize=14)
plt.ylabel('Direction [rad]',fontsize=14)
plt.title('Multi-step Direction Prediction (RMSE)',fontsize=14)
plt.legend(fontsize=14)
plt.show()
```
