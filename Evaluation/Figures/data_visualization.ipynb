```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
df = sns.load_dataset('tips')

sns.boxplot(x="day", y="total_bill", hue="smoker", data=df, palette="Set1", width=0.5)
plt.show()
```

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

path = "/home/yuchen/sftpFolder/pattern_boxplot.csv"
data = pd.read_csv(path, header=0, usecols=[0,1,2])

# Usual boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='pattern', y='vel',  hue="model", data=data, width=0.6, showmeans=True, color='skyblue')
plt.xticks(fontsize=14)  
plt.yticks(fontsize=14)
plt.xlabel('Input Pattern',fontsize=14)
plt.ylabel('Vel [mm/s]',fontsize=14)
plt.title('Single-step Velcity Prediction (RMSE)',fontsize=14)
plt.legend(fontsize=14)
plt.show()

path = "/home/yuchen/sftpFolder/pattern_boxplot.csv"
data = pd.read_csv(path, header=0, usecols=[0,1,3])
data

# Usual boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='pattern', y='direction',  hue="model", data=data, width=0.6, showmeans=True, color='skyblue')
plt.xticks(fontsize=14)  
plt.yticks(fontsize=14)
plt.xlabel('Input Pattern',fontsize=14)
plt.ylabel('Direction [rad]',fontsize=14)
plt.title('Single-step Direction Prediction (RMSE)',fontsize=14)
plt.legend(fontsize=14)
plt.show()
```
